# Exercício 02

<p style="text-align:justify;">
Crie uma classe que efetue a leitura de um valor inteiro e outro real, exibindo-os no console.
</p>


```java
import java.util.Scanner;

public class LeituraValores {
    public static void main(String[] args) {
        // Criando um objeto Scanner para ler a entrada do usuário
        Scanner scanner = new Scanner(System.in);
        
        // Pedindo e lendo um valor inteiro
        System.out.print("Digite um valor inteiro: ");
        int valorInteiro = scanner.nextInt();
        
        // Pedindo e lendo um valor real
        System.out.print("Digite um valor real: ");
        double valorReal = scanner.nextDouble();
        
        // Mostrando os valores lidos
        System.out.println("Valor inteiro digitado: " + valorInteiro);
        System.out.println("Valor real digitado: " + valorReal);
        
        scanner.close();
    }
}
```

# Exercício 03

<p style="text-align:justify;">
Crie uma classe que receba o valor de um produto e a porcentagem de desconto, calcule e mostre o valor do desconto e o valor do produto com o desconto. 
</p>

- Observação: o valor do desconto é calculado por meio da fórmula: 
    valor do desconto = valor do produto * percentual de desconto / 100.

```java
import java.util.Scanner;

public class DescontoProduto {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        
        System.out.print("Digite o valor do produto: ");
        double valorProduto = scanner.nextDouble();
        
        System.out.print("Digite a porcentagem de desconto: ");
        double percentualDesconto = scanner.nextDouble();
        
        exibirDesconto(valorProduto, percentualDesconto);
        
        scanner.close();
    }
    
    public static void exibirDesconto(double valorProduto, double percentualDesconto) {
        double valorDesconto = valorProduto * percentualDesconto / 100;
        double valorComDesconto = valorProduto - valorDesconto;
        
        System.out.println("Valor do desconto: " + valorDesconto);
        System.out.println("Valor do produto com desconto: " + valorComDesconto);
    }
}
```

# Exercício 04

<p style="text-align:justify;">
Crie uma classe que receba dois valores inteiros. 
</p>

- O primeiro valor corresponde à quantidade de pontos do líder do campeonato brasileiro de futebol. 

- O segundo valor corresponde à quantidade de pontos do time lanterna. 

<p style="text-align:justify;">
- Considerando que cada vitória vale 3 pontos, calcule o número de vitórias necessárias para que o time lanterna alcance (ou ultrapasse) o líder. 
    
- Por exemplo, supondo que as quantidades de ponto fornecidas sejam 40 e 22, então o número de vitórias apresentada na saída deverá ser 6, pois (40-22) / 3 = 6.
</p>

```java
import java.util.Scanner;

public class CalculadoraVitorias {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        
        System.out.print("Digite a quantidade de pontos do líder: ");
        int pontosLider = scanner.nextInt();
        
        System.out.print("Digite a quantidade de pontos do time lanterna: ");
        int pontosLanterna = scanner.nextInt();
        
        int vitoriasNecessarias = calcularVitoriasNecessarias(pontosLider, pontosLanterna);
        
        System.out.println("Número de vitórias necessárias para o time lanterna alcançar o líder: " + vitoriasNecessarias);
        
        scanner.close();
    }
    
    public static int calcularVitoriasNecessarias(int pontosLider, int pontosLanterna) {
        int diferencaPontos = pontosLider - pontosLanterna;
        int vitoriasNecessarias = diferencaPontos / 3;
        
        // Verifica se há sobra de pontos
        if (diferencaPontos % 3 != 0)
            vitoriasNecessarias++;
        
        return vitoriasNecessarias;
    }
}
```

# Exercício 06 
<p style="text-align:justify;">
Crie uma classe que solicite ao usuário inserir dois inteiros, obtenha dele esses números e imprima sua soma, produto, diferença e quociente (divisão).
</p>

```java
import java.util.Scanner;

public class OperacoesAritmeticas {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        
        // Solicita ao usuário que insira dois inteiros
        System.out.print("Digite o primeiro inteiro: ");
        int num1 = scanner.nextInt();
        
        System.out.print("Digite o segundo inteiro: ");
        int num2 = scanner.nextInt();
        
        // Realiza as operações aritméticas
        int soma = num1 + num2;
        int produto = num1 * num2;
        int diferenca = num1 - num2;
        // Certifique-se de lidar com divisão por zero
        double quociente = (num2 != 0) ? (double) num1 / num2 : Double.NaN;
        
        // Exibe os resultados
        System.out.println("Soma: " + soma);
        System.out.println("Produto: " + produto);
        System.out.println("Diferença: " + diferenca);
        System.out.println("Quociente: " + (Double.isNaN(quociente) ? "Indefinido (divisão por zero)" : quociente));
        
        scanner.close();
    }
}
```

# Exercício 07 

Declare uma classe para calcular o produto de três inteiros.

Siga as instruções abaixo:

1. Crie um Scanner chamado input que leia valores a partir da entrada padrão.
2. Declare as variáveis x, y, z e result como tipo int.
3. Solicite que o usuário insira o primeiro inteiro.
4. Leia o primeiro inteiro digitado pelo usuário e armazene-o na variável x.
5. Solicite que o usuário insira o segundo inteiro.
6. Leia o segundo inteiro digitado pelo usuário e armazene-o na variável y.
7. Solicite que o usuário insira o terceiro inteiro.
8. Leia o terceiro inteiro digitado pelo usuário e armazene-o na variável z.
9. Compute o produto dos três inteiros contidos nas variáveis x, y e z e atribua o resultado à variável result.
19. Use System.out.printf para exibir a mensagem “O valor do produto é” seguida pelo valor da variável result. 

```java
import java.util.Scanner;

public class CalculadoraProduto {
    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);

        // Declaração das variáveis x, y, z e result
        int x, y, z, result;

        // Solicita ao usuário que insira o primeiro inteiro
        System.out.print("Digite o primeiro inteiro: ");
        // Lê o primeiro inteiro digitado pelo usuário e armazena-o na variável x
        x = input.nextInt();

        // Solicita ao usuário que insira o segundo inteiro
        System.out.print("Digite o segundo inteiro: ");
        // Lê o segundo inteiro digitado pelo usuário e armazena-o na variável y
        y = input.nextInt();

        // Solicita ao usuário que insira o terceiro inteiro
        System.out.print("Digite o terceiro inteiro: ");
        // Lê o terceiro inteiro digitado pelo usuário e armazena-o na variável z
        z = input.nextInt();

        // Computa o produto dos três inteiros e atribui o resultado à variável result
        result = x * y * z;

        // Exibe o resultado usando System.out.printf
        System.out.printf("O valor do produto é %d%n", result);

        input.close(); 
}
```
